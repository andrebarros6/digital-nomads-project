```python
import requests
import pandas as pd

def get_current_temperature(city, api_key):
    endpoint = f'http://api.weatherbit.io/v2.0/current'

    response = requests.get(
        endpoint,
        params={
            'city': city,
            'key': api_key,
        }
    )
        
    if response.status_code != 200:
        print(f"Error fetching data for {city}: {response.status_code} {response.reason}")
        return None

    data = response.json()    
    temperature = data['data'][0]['temp']
    return temperature

API_KEY = 'e2a036dbe3c34c2395eb21cb054edf1b'

cities = [
    "Paris",
    "Madrid",
    "Barcelona",
    "Berlin",
    "Rome",
    "Vienna",
    "Hamburg",
    "Munich",
    "Milan",
    "Naples"
]

current_temperatures = {city: get_current_temperature(city, API_KEY) for city in cities}
city_temp_list = [f"{temp}°C" for city, temp in current_temperatures.items() if temp is not None]
df = pd.DataFrame(city_temp_list, columns=["City_Temperature"])

display(df)
```

```python
filtered_df['City_Temperature'] = filtered_df['City_Temperature'].combine_first(df['City_Temperature'])

filtered_df
```

```python
filtered_df["User1_Total"] = (
    (4 * filtered_df["Meal, Inexpensive Restaurant (USD)"]) + 
    (10 * filtered_df["One-way Ticket (Local Transport) (USD)"]) + 
    filtered_df["Apartment (1 bedroom) Outside of Centre (USD)"]
)

filtered_df["User2_Total"] = (
    (2 * filtered_df["Meal for 2 People, Mid-range Restaurant, Three-course (USD)"]) + 
    filtered_df["Monthly Pass (Regular Price) (USD)"] + 
    filtered_df["Apartment (1 bedroom) in City Centre (USD)"]
)

filtered_df["User3_Total"] = (
    (4 * filtered_df["Meal for 2 People, Mid-range Restaurant, Three-course (USD)"]) + 
    filtered_df["Monthly Pass (Regular Price) (USD)"] + 
    filtered_df["Monthly Pass Fitness Club"] + 
    filtered_df["Apartment (1 bedroom) in City Centre (USD)"] + 
    filtered_df["Apartment (3 bedrooms) in City Centre (USD)"]
)
filtered_df.head()
```

```python
usd_to_eur = 0.92  

columns_to_convert = [col for col in filtered_df.columns if "(USD)" in col or col.startswith("User")]

filtered_df[columns_to_convert] = filtered_df[columns_to_convert] * usd_to_eur

filtered_df.rename(columns=lambda col: col.replace("(USD)", "(EUR)"), inplace=True)

filtered_df = filtered_df.round(2)

filtered_df
```

```python
selected_columns = ['city', 
                    'remote connection: average wifi speed (mbps per second)', 
                    'co-working spaces: number of co-working spaces', 
                    'after-work drinks: average price for 2 beers in a bar']

spaces_filter = filtered_df[[col for col in selected_columns if col in filtered_df.columns]]
spaces_filter
```
